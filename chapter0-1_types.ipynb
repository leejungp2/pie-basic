**4) 인덱싱과 슬라이싱**
> 인덱싱: 문자열 내부의 문자 '하나'를 선택하는 연산자

>'문자열[인덱스]'로 호출

- 대괄호 [ ] 안에는 선택할 문자의 위치를 지정하며, 이 숫자를 인덱스(index)라고 부름
    - 순방향 인덱스: 0부터 시작
    - 역방향 인덱스: -1부터 시작

```python
# 문자열 인덱싱 : 문자 하나 선택
#           012   3  4567  8 910111213
#                                 -2-1
sentence = '파이썬 하루만에 끝내보자!'
```

```python
# 문자열[인덱스]로 호출
# 순방향 인덱스 : 0 부터 시작
# 역방향 인덱스 : -1 부터 시작
print(sentence)
print(sentence[0])
print(sentence[-1])
```

**4) 인덱싱과 슬라이싱**
> 슬라이싱: 문자열의 특정 '범위'를 선택할 때 사용하는 연산자

> 문자열[인덱스1 : 인덱스2]로 호출

- 파이썬은 '마지막 숫자를 포함하지 않음'으로 적용
    - ex) 인덱스 1이상 ~ 인덱스 2미만

```python
# 문자열 슬라이싱 : 특정 범위 선택
#           012   3  4567  8 910111213
#                                 -2-1
sentence = '파이썬 하루만에 끝내보자!'
```

```python
# 문자열[인덱스1 : 인덱스2]로 호출
# 4:8 = 4, 5, 6, 7
# 4 이상부터 8 미만까지의 범위 선택하여 출력
# 마지막 숫자는 포함하지 않음
print(sentence)
print(sentence[4:8])
```

**5) 포맷팅**
> 개념: 문자열 안에 어떤 값을 삽입하는 방법

- 문자열 포맷 코드에 맞춰 바로 숫자와 문자를 대입할 때 사용

**(1) %연산자 포맷팅**
- 숫자(정수) 대입: %d

- 숫자(실수) 대입: %f

- 문자 1개 대입: %c

- 문자열 대입: %s

```python
# 한 개의 % 연산자 포맷팅
# 숫자 대입은 %d
sentence = "나는 %d호선을 타고 다녀." % 6
print(sentence) 
```

```python
# 두 개 이상의 % 연산자 포맷팅
# 숫자 나타내는 변수와 문자를 나타내는 변수 대입
# 숫자 대입은 %d, 문자 대입은 %s
a = 4
b = "학교"
sentence = "나는 %d호선을 타고 %s를 가." % (a, b)
print(sentence)
```

**(2) format() 함수 포맷팅**
- 중괄호{ }를 포함한 문자열 뒤에 도트(.)를 찍고 format( ) 함수 사용
- 중괄호 개수와 format( ) 함수 괄호 안의 매개변수 개수는 동일
- 함수를 사용하면 문자열의 {} 기호가 format( ) 함수 괄호 안에 있는 매개변수로 차례로 대치

```python
# 한 개의 format() 함수 포맷팅
# { } 안에 숫자 대입
# format 함수 안에 넣을 숫자 삽입
print("나는 {}호선을 타고 다녀.".format(6))
print("나는 {0}호선을 타고 다녀.".format(6))
```

**6)관련 함수**
- len(x) : 문자열 길이 출력
- split() : 문자열 나누기
- count(x) : 문자 개수 세기
- replace('a', 'b') : 문자열 바꾸기
- find('a') : 위치 알려주기
- upper() : 소문자를 대문자로 바꾸기
- lower() : 대문자를 소문자로 바꾸기
- join(x) : 문자열 삽입

```python
## 함수는 뒤에 .형태로 붙는거구나.
# len(x) : 문자열 길이 출력
sentence1 = "뽀로로는 노는 게 제일 좋아."
print('len(x) 문자열 길이: ', len(sentence1))
print(len(sentence1))

# split() : 문자열 나누기
print('split(x) 문자열 나누기: ', sentence1.split())

# count(x) : 문자 개수 세기
print('count(x) 문자 개수 세기:', sentence1.count('로'))

# replace('a', 'b') : 문자열 바꾸기
sentence2 = sentence1.replace('노는', "먹는")
print(sentence1)
print('replace() 문자열 바꾸기: ', sentence2)

# find('a') : 위치 알려주기
file_name = 'HAMburger.jpg'
print('find(\'a\') 위치 알려주기: ', file_name.find('.'))
print(file_name.split('.'))


# upper() : 소문자를 대문자로 바꾸기
print('upper() 소->대: ', file_name.upper())

# lower() : 대문자를 소문자로 바꾸기
print('lower() 대->소: ', file_name.lower())

# join(x) : 문자열 삽입
sentence = '.'.join('abcde')
print('join(x) 문자열 삽입: ', sentence)
```
