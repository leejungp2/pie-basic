# 2. 매개변수(parameter) (3)
- 개념: 함수에 입력으로 전달된 값을 받는 변수 (함수를 정의하는 측)

## 2-1. 디폴트 매개변수
- 매개변수의 값이 입력되지 않으면 디폴트 값으로 자동입력(에러가 나지 x)
- 매개변수가 여러 개인 경우 뒤의 매개변수부터 디폴트 값을 가질 수 있음(인수는 왼쪽부터 차례대로 넘겨줌)

        def 함수이름(매개변수 = 디폴트 값):
        수행할 문장1
        return 결과

```python
#디폴트 매개변수 예시
def add(x, y =3):
    return x + y
```

```python
#y 매개변수에 값을 입력하면 입력값대로 출력
z = add(3, 6)
print(z)
```

```python
#y 매개변수에 입력이 없다면 디폴트 값으로 출력
z = add(3)
print(z)
```

```python
#cf) 일반 함수 예시
## 항상 두 개의 인수를 받아야 하며, 하나라도 받지 않을 시 에러 발생
def greet(name, hello):
    print("Hi", name,',', hello)

greet('Jung', 'good morning!')
```

## 2-2. 가변 매개변수
- 개념: 원하는 만큼의 인자를 받을 수 있는 함수
- 표현: 매개변수 앞에 *(튜플 형식) 또는 **(딕셔너리 형식)를 붙임
    - *args
    - **kwargs
- 생성된 가변 매개변수는 리스트처럼 사용
- 제약:
    - 뒤에 일반 매개변수가 올 수 없음
    - 하나만 사용할 수 있음

        def 함수이름(*매개변수):
        수행할 문장1
        return 결과

```python
#예제 1
def menu(*values):
    print("<오늘 점심 메뉴 추천>")

    for value in values:
        print(value)
```

```python
m = menu('김밥', '짜장면', '파스타', '샐러드')
```

```python
#예제2
def para_func(*para):
    result = 0
    for num in para:
        result = result + num
    
    return result

# 전역변수 선언
hap = 0

#메인 코드
hap = para_func(10, 20)
print("매개변수가 2개인 함수를 호출한 결과 ==> %d" % hap)
hap = para_func(10, 20, 30)
print("매개변수가 3개인 함수를 호출한 결과 ==> %d" % hap)
```

```python
#예제3
## s문자열 d정수형 f실수형
def dic_func(**para) :
    for k in para.keys():
        print(f"{k} --> {para[k]}명입니다.")

dic_func(트와이스 = 9, 소녀시대 = 7, 걸스데이 = 4, 블랙핑크 = 4)
```

            return 되돌려주는 결과값
        
        변수 = 함수이름(argument)

```python
#두 개의 리턴 값
def returnTest(): #기본으로 반환하는 매개변수 없음
    a = 3
    b = 4 #함수 안에서 a, b값 고정적으로 정의. 외부에서 값 조정 불가
    return a, b

value1, value2 = returnTest()
print(value1)
print(value2)
```

# 5. 함수의 호출 (cf)

- 일반적인 호출 : Call by Value, Call by Reference
- 파이썬 : Call by Object Reference 혹은 Call by Sharing
- 모든 인수(argument)는 인수자체가 다른 객체로 치환 될때, 함수를 호출한 측에 아무런 영향을 미치지 못 한다.
- 변경 가능한 인수는 참조(Reference)를 이용하여 내부 객체를 변경 할 때 변경이 호출한 측에 반영 된다.
- 함수 안에서는 정의되지 않은 다른 함수를 호출하여도 된다.

| 자료형 | 설명 | 예|
|------|------|-----|
|변경 가능형 (Mutable) | 데이터의 값을 변경할 수 있다.| list, dict, set|
|변경 불가능형 (Immumtable) | 데이터의 값을 변경할 수 없다. | int, float, complex, str, tuple, frozenset|

### 튜플과 사전을 사용한 함수 호출
- 함수 호출에 사용하는 인수(argument)들이 튜플이면 *를 이용하여 함수를 호출할 수 있다.
- 함수 호출에 사용하는 인수(argument)들이 사전이면 **를 이용하여 함수를 호출할 수 있다.

```python
def func_f(a, b, c):
    print(a, b, c)

a = (1, 2, 3)
```

```python
func_f(a[0], a[1], a[2])
```

```python
#튜플
func_f(*a)
```

```python
dict_args = {'a':1, 'b':2, 'c':3}
```

```python
#딕셔너리
func_f(**dict_args)
```

# 6. 유효 범위 규칙(scope rule)

<파이썬 3 바이블, 이강성, 프리렉>

- 개념:유효하게 사용되는 문맥(context) 범위를 정하는 규칙
- 변수가 특정 범위에서 유효한지를 결정

## LEGB

파이썬에서 변수에 값을 바인딩하거나 변수의 값을 참조하는 경우 LEGB 규칙을 따릅니다.

LEGB는 각각 다음과 같은 의미를 가집니다.


L -	Local scope의 약자로 함수 안을 의미합니다.

E -	Enclosed scope의 약자로 내부함수에서 자신의 외부 함수의 범위를 의미합니다.

G -	Global scope의 약자로 함수 바깥 즉, 모듈 범위입니다.
- 전역변수(global varaible)

B -	Built-in scope의 약자로 open, range와 같은 파이썬 내장 함수들을 의미합니다.


scope_rule.png

```python
#예제
import random

#함수 선언
def getNumber():
    return random.randrage(1, 16)

#전역 변수 선언
lotto = []
num = 0

#메인 코드 부분
print("로또 추첨을 시작합니다.")

while True:
    num = getNumber()

    if lotto.count(num) == 0:
        lotto.append(num)

    if len(lotto) >= 6:
        break

print("추첨된 로또 번호 ==> ", end = '')
lotto.sort()
for i in range(0, 6):
    print("%d" % lotto[i], end = "")

    
```
